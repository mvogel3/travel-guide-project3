```python
import pandas as pd
```

```python
all_hotels_df = pd.read_csv('../travel-guide-project3/merged_h_avg_rat_dist_transp_ent_df.csv')
all_hotels_df
```

```python
print(all_hotels_df['Hotel Rating'])
```

```python
# Extract the rating (first number) and number of reviews (first number in parentheses) into separate columns
all_hotels_df["Rating (out of 10)"] = all_hotels_df['Hotel Rating'].str.extract(r'^([\d.]+)')
all_hotels_df['Number of Reviews'] = all_hotels_df['Hotel Rating'].str.extract(r'\(([\d,]+)')
all_hotels_df
```

```python
all_hotels_df.info()
```

```python
all_hotels_df['Rating (out of 10)'] = all_hotels_df['Rating (out of 10)'].astype(float)


import pandas as pd

# Clean the 'Number of Reviews' column by removing non-digit characters and commas
all_hotels_df['Number of Reviews'] = all_hotels_df['Number of Reviews'].str.replace('[^\d]', '').str.replace(',', '')

# Convert the 'Number of Reviews' column to float
all_hotels_df['Number of Reviews'] = pd.to_numeric(all_hotels_df['Number of Reviews'], errors='coerce').astype(float)

all_hotels_df.info()
```

```python
all_hotels_df = all_hotels_df.rename(columns={"Hotel Name": "hotel_name",
                                             "Hotel Number": "hotel_number",
                                             "Hotel Street": "hotel_street",
                                             "City": "city",
                                             "Hotel Full Address": "hotel_full_address",
                                             "Advertised Price": "advertised_price",
                                             "Total Price": "total_price",
                                             "Hotel Rating": "hotel_rating",
                                             "Distance from NYC": "distance_from_nyc",
                                             "Hotel Longitude": "hotel_longitude",
                                             "Hotel Latitude": "hotel_latitude",
                                             "Transportation Longitude": "transportation_longitude",
                                             "Transportation Latitude": "transportation_latitude",
                                             "Entertainment": "entertainment_place",
                                             "Entertainment Full Address": "ent_full_address",
                                             "Entertainment Longitude": "ent_longitude",
                                             "Entertainment Latitude": "ent_latitude",
                                             "Rating (out of 10)": "rating_out_of_10",
                                             "Number of Reviews": "no_of_reviews"})
all_hotels_df
```
