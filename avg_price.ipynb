```python
price_csv=["April5_expedia.csv", "April12_expedia.csv", "April19_expedia.csv", "April26_expedia.csv", "August4_expedia.csv", "August11_expedia.csv", "August18_expedia.csv", "August25_expedia.csv", "December1_expedia.csv", "December8_expedia.csv", "December15_expedia.csv", "December22_expedia.csv", "December29_expedia.csv", "February2_expedia.csv", "February9_expedia.csv", "February16_expedia.csv", "February23_expedia.csv", "January5_expedia.csv", "January12_expedia.csv", "January19_expedia.csv", "January26_expedia.csv", "July7_expedia.csv", "July14_expedia.csv", "July21_expedia.csv", "July28_expedia.csv", "June2_expedia.csv", "June9_expedia.csv", "June16_expedia.csv", "June23_expedia.csv", "June30_expedia.csv", "March1_expedia.csv", "March8_expedia.csv", "March15_expedia.csv", "March22_expedia.csv", "March29_expedia.csv", "May3_expedia.csv", "May10_expedia.csv", "May17_expedia.csv", "May24_expedia.csv", "November3_expedia.csv", "November10_expedia.csv", "November17_expedia.csv", "November24_expedia.csv", "October6_expedia.csv", "October13_expedia.csv", "October20_expedia.csv", "October27_expedia.csv", "September1_expedia.csv", "September8_expedia.csv", "September15_expedia.csv", "September22_expedia.csv", "September29_expedia.csv"]

price_df = df = pd.DataFrame()
for file in price_csv:
    week_price_df = pd.read_csv(f"HotelData/{file}", encoding="utf-8")
    price_df = price_df.append(week_price_df, ignore_index=True)
```

```python
price_df['Advertised Price'] = price_df['Advertised Price'].str.extract(pat='(\d+)', expand=False)
price_df['Total Price'] = price_df['Total Price'].str.extract(pat='(\d+)', expand=False)
price_df.head()
```

```python
price_df = price_df.dropna()
price_df=price_df.astype({'Advertised Price': 'float64', 'Total Price': 'float64'})
price_df=price_df.sort_values(by=['Hotel Name'])
price_df
```

```python
price_df_avg = price_df.groupby('Hotel Name').mean()
price_df_avg=price_df_avg.sort_values(by=['Hotel Name'])
price_df_avg
```

```python
price_df_avg.to_csv("aggregated/avg_price.csv")
```
